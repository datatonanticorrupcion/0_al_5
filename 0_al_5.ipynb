De lo analizado previamente se pueden observar varios aspectos de interes: 
  * Impuesto Especial sobre Produccion y Servicios (IEPS): hay muchos datos negativos, incluso el primer cuartil es negativo, lo que significa que 25 % de los meses (aproximadamente 28 meses) hubo mayor devolucion que recaudacion, lo cual es algo relevante a considerar.
  * El componente principal del IEPS es el impuesto a gasolina y diesel, en donde incluso la mediana es negativa lo que significa que en mas de la mitad de los meses (aproximadamente 55) hubo mas devolucion que recaudacion. 
  * En todos los tipos de impuestos la recaudacion es positiva menos: Aguas, refrescos y sus concentrados (IEPS)       y Otros (IEPS), donde la suma de recaudacion registrada es cero, lo que significa que en terminos reales no se aplica este impuesto.
  * Se puede apreciar que el impuesto que mas aporte a la nacion es el impuesto sobre la renta, seguido por el impuesto al valor agregado y por el IEPS en su rubo de gasolinas y diesel.
  * Los demas impuestos relacionados al IEPS son muy marginales.
  * La variacion de la recaudacion en cada uno de los rubros es grande, por lo que se puede pensar o formular una hipotesis que la economia ha crecido en terminos generales, o bien, hay muchisima variacion por temporada en la contribucion de impuestos. El tipo de impuesto con mayor variacion es el impuesto sobre la renta asociada principalmente a los impuestos que pagan los empleados de su sueldo, por lo que se cree que una hipotesis coherente es que la economia haya crecido. Dada esta hipotesis y que en los impuestos del IEPS hay muchos numeris negativos, se realiza un analisis temporal.

```python
impuestos1.plot(figsize=(15,15))
```

```python
impuestos1['Total de Ingresos tributarios'].pct_change(periods=12).mean()
```

De la grafica anterior se puede ver que la recaudacion total tiene una tendencia clara a la alza (en promedio 9.8 % anual) mayor que el crecimiento economico (2.5 % durante sexenio de EPN), principalmente ocasionada por el incremento en el impuesto sobre la renta y el iva lo que implica que la gente en terminos reales tiene menos dinero para gastar de manera individual. El IEPS hasta antes del mes aproximadamente 45 oscilaba entre ligeramente positivo y negativo y que a partir de este momento tuvo un incremento importante y que se ha mantenido hasta la fecha, aunque ya medianamente se estabilizo, alcanzando su punto maximo en:
